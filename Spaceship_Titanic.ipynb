## **Standardization**

```python
from sklearn.preprocessing import StandardScaler
scaler_spaceship = StandardScaler()
X_spaceship = scaler_spaceship.fit_transform(X_spaceship)
```

```python
X_spaceship
```

## **Training base and test base**

```python
from sklearn.model_selection import train_test_split
```

```python
X_spaceship_training, X_spaceship_test, Y_spaceship_training, Y_spaceship_test = train_test_split(X_spaceship, Y_spaceship, test_size = 0.25, random_state = 0)
```

```python
import pickle
```

```python
with open('spaceship.pkl', mode = 'wb') as f:
  pickle.dump([X_spaceship_training, X_spaceship_test, Y_spaceship_training, Y_spaceship_test], f)
```

```python
with open('spaceship.pkl', 'rb') as f:
  X_spaceship_training, X_spaceship_test, Y_spaceship_training, Y_spaceship_test = pickle.load(f)
```

## **Naïve Bayes**

```python
from sklearn.naive_bayes import GaussianNB
```

```python
naive_spaceship=GaussianNB()
```

```python
naive_spaceship.fit(X_spaceship_test, Y_spaceship_test)
```

```python
predictions=naive_spaceship.predict(X_spaceship_test)
```

```python
predictions
```

```python
Y_spaceship_test
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(Y_spaceship_test, predictions)
```

```python
print(classification_report(Y_spaceship_test, predictions))
```

```python
confusion_matrix(Y_spaceship_test, predictions)
```

```python
from yellowbrick.classifier import ConfusionMatrix
```

```python
cm = ConfusionMatrix(naive_spaceship)
cm.fit(X_spaceship_training, Y_spaceship_training)
cm.score(X_spaceship_test, Y_spaceship_test)
```
